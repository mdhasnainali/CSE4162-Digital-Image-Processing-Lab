### 1. Take grayscale image of size 512x512 and perform the following operations - 
###  (d) Illustrate the histogram of the image and make single threshold segmentation observed from the histogram

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./../../Images/Skeleton 750x1482.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title(f'Image Size: {image.shape}')
```

```python
def generate_histogram(image):
    height, weight = image.shape
    gray_levels_count = np.zeros(256)

    for row in range(height):
        for column in range(weight):
            gray_levels_count[image[row,column]] += 1
    
    plt.bar(range(256), gray_levels_count, width=1, color='gray')

    # OR
    # plt.hist(img.ravel(),bins=256,range=(0,256), color='gray')

    plt.title("Histogram")
    plt.show()
```

```python
generate_histogram(image)
```

```python
threshold_value = 27

segmented_image = np.where(image < threshold_value, 0, 255)

plt.imshow(segmented_image, cmap='gray')
plt.title("Segmented Image")
plt.show()
```

```python
generate_histogram(segmented_image)
```
