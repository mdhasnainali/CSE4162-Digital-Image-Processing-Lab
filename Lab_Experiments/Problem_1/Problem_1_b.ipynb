### 1. Take grayscale image of size 512x512 and perform the following operations - 
### (b) Decrease its spatial resolution by half every time and observe its change when displaying in the same window size

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./../../Images/Rose 1024x1024.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title(f'Image Size: {image.shape}')
```

#### Spatial Resolution Decreasing Function

```python
def decrease_resolution(image):
    height, width = image.shape
    n_image = np.zeros((height//2, width//2))    
    for row in range(height):
        for column in range(width):
            n_image[row//2, column//2] = image[row,column]
    return n_image
```

#### Display the changes of resolution

```python
decrease_image = cv2.resize(image, (512,512))
plt.figure(figsize=(10,10))
for i in range(1,5):
    plt.subplot(2,2, i)
    decrease_image = decrease_resolution(decrease_image)
    plt.imshow(decrease_image, cmap='gray')
    plt.title(f'Image Size: {decrease_image.shape}')  

plt.show()  
```
