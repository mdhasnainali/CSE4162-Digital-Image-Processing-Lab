### 1. Take grayscale image of size 512x512 and perform the following operations - 
### (c) Decrease it intensity level resolution by one bit up to reach its binary format observe its change when displaying in the same window size

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./../../Images/Skull 374x452.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title(f'Image Size: {image.shape}')
```

```python
decrease_image = image.copy()
plt.figure(figsize=(20,10))
for i in range(1,9):
    plt.subplot(2,4, i)
    plt.imshow(decrease_image, cmap='gray')
    plt.title(f'Image: {i-}bit')  
    decrease_image = decrease_image//2

plt.show()  
```
