### 1. Take grayscale image of size 512x512 and perform the following operations - 
### (a)  Decrease its spatial resolution by half

```python
import cv2 
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./../../Images/Rose 1024x1024.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.title(f'Image Size: {image.shape}')
```

```python
def decrease_resolution(image):
    height, width = image.shape
    n_image = np.zeros((height//2, width//2))    
    for row in range(height):
        for column in range(width):
            n_image[row//2, column//2] = image[row,column]
    return n_image
```

```python
new_image = decrease_resolution(image)
plt.imshow(new_image, cmap='gray')
plt.title(f'Image Size: {new_image.shape}')
```
