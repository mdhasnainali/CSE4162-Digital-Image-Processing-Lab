### (c)  Find the difference image between original and the image obtained by last three bits (MSB)

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('./../../Images/Dollar 1192x500.tif', cv2.IMREAD_GRAYSCALE)
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title("The 100 Dollar Image")
plt.show()
```

```python
mask = "11100000"
three_bit_image = image & int(mask, 2)

plt.imshow(cv2.cvtColor(three_bit_image, cv2.COLOR_BGR2RGB))
plt.title("3 Bit Dollar Image")
plt.show()
```

```python
difference_image = cv2.absdiff(image, three_bit_image)
plt.imshow(difference_image,cmap='gray')
plt.title("The Difference Image")
plt.show()
```
