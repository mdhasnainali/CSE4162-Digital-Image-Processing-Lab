### Take grayscale image of size 512x512 and perform the following operations â€“ 
### (a)  Perform the brightness enhancement of a specific range of gray levels and observe its result

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
image = cv2.imread("./../../Images/Fractured_Spine_746x976.tif", cv2.IMREAD_GRAYSCALE)
plt.imshow(image, cmap='gray')
plt.show()
```

```python
start, end, add_value = 10, 40, 50
height, weight = image.shape

for row in range(height):
    for column in range(weight):
        if start <= image[row,column] <= end:
            image[row,column] += add_value
        
        image[row,column] = 255 if image[row,column] > 255 else image[row,column] 

plt.imshow(image, cmap='gray')
plt.show()
```
