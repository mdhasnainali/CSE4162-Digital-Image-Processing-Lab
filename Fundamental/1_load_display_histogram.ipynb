### Image Load and Display

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

### Basic Image

```python
image = cv2.imread('../Images/four.png')
image = cv2.resize(image, (8,8), interpolation = cv2.INTER_LINEAR)
image = image[:,:,1]
plt.imshow(image, cmap="gray")
plt.show()
```

```python
image.shape
```

```python
image
```

```python
image_numpy = np.array(image)
image_numpy.shape
```

```python
values, counts = np.unique(image_numpy, return_counts=True)
values, counts
```

```python
# Histogram
plt.hist(image_numpy.reshape(-1))
```
